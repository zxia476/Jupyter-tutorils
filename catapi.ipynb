```python

cols = ['name', 'affection_level']


bar_data = pd.DataFrame(df[cols])

# Sort the DataFrame by affection level
bar_data = bar_data.sort_values(by=['affection_level'])

# Create the bar plot
plt.bar(bar_data['name'], bar_data['affection_level'])

# Set the title and axis labels
plt.title('Bar Plot of Affection Level and Name')
plt.xlabel('Name')
plt.ylabel('Affection Level')

# Rotate the x-axis labels for readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python

cols = ['name', 'intelligence']


bar_data = pd.DataFrame(df[cols])

# Sort the DataFrame by affection level
bar_data = bar_data.sort_values(by=['intelligence'])

# Create the bar plot
plt.bar(bar_data['name'], bar_data['intelligence'])

# Set the title and axis labels
plt.title('Bar Plot of intelligence and Name')
plt.xlabel('Name')
plt.ylabel('Affection Level')

# Rotate the x-axis labels for readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
cols = ['health_issues', 'intelligence']

# Create a new DataFrame with only the selected columns
scatter_data = pd.DataFrame(df[cols])

# Create the scatter plot
plt.scatter(scatter_data['health_issues'], scatter_data['intelligence'])

# Set the title and axis labels
plt.title('Scatter Plot of health and intelligence')
plt.xlabel('health')
plt.ylabel('intelligence')

# Show the plot
plt.show()
```
